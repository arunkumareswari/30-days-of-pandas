**Problem Name: Recyclable and Low Fat Products**

```python
import pandas as pd
```

### Create DataFrame

```python
data = [
    ['0', 'Y', 'N'],
    ['1', 'Y', 'Y'], 
    ['2', 'N', 'Y'], 
    ['3', 'Y', 'Y'], 
    ['4', 'N', 'N']
    ]

products = pd.DataFrame(
    data, 
    columns=['product_id', 'low_fats', 'recyclable']
)
```

```python
(products['low_fats'] == 'Y') & (products['recyclable'] == 'Y')
```

```python
products[ (products['low_fats'] == 'Y') & (products['recyclable'] == 'Y')]
```

```python
### Define the solution class 

def find_products(products: pd.DataFrame) -> pd.DataFrame:

    df = products[ (products['low_fats'] == 'Y') & (products['recyclable'] == 'Y')]

    return df[['product_id']]

    
```

```python
result = find_products(products)
result
```
