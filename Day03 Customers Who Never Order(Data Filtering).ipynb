**Problem Name: Customers Who Never Order**

```python
# Create 

import pandas as pd

data = [
    [1, 'Joe'], 
    [2, 'Henry'], 
    [3, 'Sam'], 
    [4, 'Max']
    ]
    
customers = pd.DataFrame(
    data, 
    columns=['id', 'name']
    ).astype({'id':'Int64', 'name':'object'})

data = [
    [1, 3], 
    [2, 1]
    ]
orders = pd.DataFrame(
    data, 
    columns=['id', 'customerId']
    ).astype({'id':'Int64', 'customerId':'Int64'})
```

```python

customers
```

```python
orders
```

```python
# Left join customers + orders

merged = pd.merge(customers, orders, left_on = 'id', right_on = 'customerId', how = 'left')
merged
```

```python
# check null values
merged['customerId'].isna()
```

```python
# Filters customers with No orders

merged = merged[merged['customerId'].isna()]  # never order customer
merged
```

```python
# Select only name and rename column

merged[['name']].rename(columns= {'name':'Customers'})
```

```python


def find_customers(customers : pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:

    merged = pd.merge(customers, orders, left_on = 'id', right_on = 'customerId', how = 'left')
    merged = merged[merged['customerId'].isna()]

    return merged[['name']].rename(columns= {'name': 'Customers'})

```

```python
result = find_customers(customers, orders)
result
```
